```python
import wmi
import time
import pyautogui
import shutil
import os
import socket
import subprocess
import win32wnet
import win32netcon
import pandas as pd
import math
from openpyxl import load_workbook
from datetime import datetime

table = pd.read_excel('ComputerList.xlsx')
table.set_index('RowID',inplace=True)
TotalRow=len(table)+1
excel_file = load_workbook('ComputerList.xlsx')
report_sheet = excel_file['ComputerSheet']
user_username ="USERNAME"
user_password = "PASSWORD" 

if os.path.exists("results.txt"):
    os.remove("results.txt")
else:
    print ("The file does not exist")

results_file = open("results.txt","w")
ComputerNameList_list = []

while TotalRow > 0 :
    Record = (f"b{TotalRow}")
    RecordValue = report_sheet[f'{Record}'].value
    print (f'{TotalRow} {RecordValue}')
    ComputerNameList_list.append(RecordValue)
    #print(ComputerNameList_list)
    TotalRow = TotalRow - 1
#print (ComputerNameList_list)
i=1
for ComputerName in ComputerNameList_list:
    response = os.popen(f"ping {ComputerName} -n 1").read()
    if "Destination host unreachable." in response:
        print (f"{i} {ComputerName}:Ping Unreachable" + "\n")
        now = datetime.now()
        current_time = now.strftime("%H:%M:%S")
        results_file.write(f"{current_time}::Down::{ComputerName}::Ping Unreachable::::" + "\n")
    elif "Received = 1" and "Approximate" in response:
        ip_address = socket.gethostbyname(ComputerName)
        print (f"{i} {ComputerName}:Ping Successful:{ip_address}" + "\n")
        now = datetime.now()
        current_time = now.strftime("%H:%M:%S")
        try:
            connection = wmi.WMI(ComputerName, user = user_username, password = user_password)
            for info in connection.Win32_LogicalDisk():
                #print(info.Caption, math.ceil(int(info.FreeSpace) / (2**30)), math.ceil(int(info.Size) / (2**30)))
                #print("info-Caption = ", info.Caption, "info-FreeSpace = ", math.ceil(int(info.FreeSpace) / (2**30)), "info-Size = ", math.ceil(int(info.Size) / (2**30)))
                if info.Caption == "C:":
                    C_Drive_Space =(";info-Caption =; ", info.Caption, ";info-FreeSpace =; ", math.ceil(int(info.FreeSpace) / (2**30)), ";info-Size =; ", math.ceil(int(info.Size) / (2**30)))
                    #print (C_Drive_Space)
            for os in connection.Win32_OperatingSystem():
                computer_os = (os.name,os.osarchitecture)        
                #print (os.name,os.osarchitecture)
                machineinfo = (C_Drive_Space," ; ",computer_os)
                print (f'Machine Info : {machineinfo}')
            results_file.write(f"{current_time}::Up::{ComputerName}::Ping Successful:{ip_address}::{machineinfo}::" + "\n")
            import os
        except:
            machineinfo = "Cannot Get Info-RPC Server Unavail"
            print (machineinfo)
            results_file.write(f"{current_time}::Up::{ComputerName}::Ping Successful:{ip_address}::{machineinfo}::" + "\n")
            import os            
    else:
        print (f"{i} {ComputerName}:Ping Unsuccessful" + "\n")
        now = datetime.now()
        current_time = now.strftime("%H:%M:%S")
        results_file.write(f"{current_time}::Down::{ComputerName}::Ping Unsuccessful::::" + "\n")
        
    i=i+1
results_file.close()

```
