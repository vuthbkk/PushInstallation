```python
import time
import pyautogui
import os
import subprocess
from subprocess import PIPE, Popen
import shutil
import win32wnet
import win32netcon
import pandas as pd
from openpyxl import load_workbook

table = pd.read_excel('ComputerListToRunCmd.xlsx')
table.set_index('RowID',inplace=True)
TotalRow=len(table)+1
excel_file = load_workbook('ComputerListToRunCmd.xlsx')
report_sheet = excel_file['ComputerSheet']
print (f"Check Point Pre 1 - Records to run is {TotalRow} rows.")

ComputerNameList_list = []
while TotalRow > 0 :
    Record = (f"b{TotalRow}")
    RecordValue = report_sheet[f'{Record}'].value
    ComputerNameList_list.append(RecordValue)
    TotalRow = TotalRow - 1

print (f"Check Point Pre 2 -  {ComputerNameList_list} openned.")
results_file = open("resultsrun.txt","w")
print (f"Check Point Pre 3 -  resultsrun.txt  openned.")
       
i=0

for ComputerName in ComputerNameList_list:
    print (f"Check Point 1 - get target computer i = {i}")
    nr = win32wnet.NETRESOURCE()
    nr.dwScope = win32netcon.RESOURCE_GLOBALNET
    nr.dwType = win32netcon.RESOURCETYPE_DISK
    nr.dwUsage = win32netcon.RESOURCEUSAGE_CONNECTABLE
    nr.lpLocalName = 'p:'
    RemoteName = (ComputerNameList_list[i])
    username ="USERNAME"
    password = "PASSWORD"
    
    time.sleep(1)
    os.system("start cmd /k")
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.typewrite(f'cmdkey /add:{ComputerName} /user:{username} /pass:{password}')
    #pyautogui.typewrite('dir/w')
    time.sleep(1)
    pyautogui.press('enter')
    pyautogui.typewrite('exit')
    time.sleep(1)
    pyautogui.press('enter')

    print (f'Check Point 2 -Creating PopenProcess for {ComputerName}')
    popenprocess = f'PsExec_2.20.exe \\\\{ComputerName} -s cmd'
    p = Popen(popenprocess, shell=True, stdout=PIPE, stderr=PIPE)
    print (f'Check Point 3 - To run {p} ')
    stdout, stderr = p.communicate()
    time.sleep(1)
    response = stderr.decode('UTF-8')
    time.sleep(1)
    print (f'Check Point 4 - result = >>>')
    print (response)
    time.sleep(1)

    #if 1 > 1:
    if "Could not start PSESESVC service." and "Access is denied" in response:
        print (f"Check Point 5A -{ComputerName}:Cound not start" + "\n")
        results_file.write(f"{ComputerName}:Could not start::" + "\n")
        time.sleep(1)
        os.system("start cmd /k")
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(1)
        pyautogui.typewrite(f'cmdkey /delete:{ComputerName}')
        #pyautogui.typewrite('dir/w')
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(1)
        pyautogui.typewrite('exit')
        time.sleep(1)
        pyautogui.press('enter')
       
    elif "The handle is invalid" and "Couldn't access"in response:
        print (f"Check Point 5B -{ComputerName}:The handle is invalid" + "\n")
        results_file.write(f"{ComputerName}:The handle is invalid::" + "\n")
        time.sleep(1)
        os.system("start cmd /k")
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(1)
        pyautogui.typewrite(f'cmdkey /delete:{ComputerName}')
        #pyautogui.typewrite('dir/w')
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(1)
        pyautogui.typewrite('exit')
        time.sleep(1)
        pyautogui.press('enter')
        
    else:
        pass
        print (f"Check Point 5C -{ComputerName}:Startting PsExec" + "\n")
        time.sleep(1)
        os.system("start cmd /k")
        pyautogui.press('enter')
        pyautogui.typewrite('d:')
        pyautogui.press('enter')
        pyautogui.typewrite(r'cd\\')
        pyautogui.press('backspace')
        pyautogui.press('enter')
        time.sleep(2)
        pyautogui.typewrite(f'PsExec_2.20.exe \\\{ComputerName} -s cmd')
        pyautogui.press('enter')
        time.sleep(2)
        pyautogui.typewrite(r'cd..')
        pyautogui.press('enter')
        time.sleep(2)
        pyautogui.typewrite('cd Agentsetup')
        pyautogui.press('enter')
        time.sleep(2)
        pyautogui.typewrite('ResetGUID_SCCM.bat')
        pyautogui.press('enter')
        time.sleep(40)
        pyautogui.press('enter')
        pyautogui.typewrite('ccmsetup.bat')
        pyautogui.press('enter')
        time.sleep(15)
        pyautogui.press('enter')
        time.sleep(2)
        pyautogui.typewrite('exit')
        pyautogui.press('enter')
        time.sleep(2)
        pyautogui.typewrite('exit')
        pyautogui.press('enter')
        time.sleep(1)
        os.system("start cmd /k")
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(1)
        pyautogui.typewrite(f'cmdkey /delete:{ComputerName}')   
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(1)
        pyautogui.typewrite('exit')
        time.sleep(1)
        pyautogui.press('enter')        
        print (f"Check Point 6 - {ComputerName}:PSexec Started" + "\n")
        results_file.write(f"{ComputerName}:Completely Run Psexec to reset guid and reinstall::" + "\n")
        
    i = i+1
    
print ("End of Script")
results_file.close()



```
